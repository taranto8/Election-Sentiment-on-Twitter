```python
import csv

ctweets_df = pd.read_csv("cleaned new3 tweets partially4001 AFFILIATED.csv")
tweets_df = pd.read_csv("uncleaned but ... removed new3 tweets partially4001 AFFILIATED.csv")

del ctweets_df['Unnamed: 0']
del tweets_df['Unnamed: 0']
tweets_df.info()
```

```python

for i in range(4001):
    
    # this variable checks to see if we've seen the text at location i before:
    y = 0
    
    for j in range(i):
        if tweets_df.loc[i].text == tweets_df.loc[j].text:
            tweets_df.loc[i, "Affiliation"] = tweets_df.loc[j].Affiliation
            ctweets_df.loc[i, "Affiliation"] = ctweets_df.loc[j].Affiliation
            y = 1
            break 
            
    if y == 1:
        continue
        
    print("i =", i)
    print()
    print(tweets_df.loc[i].text)
    print()
    print("------------------------------------------------------------")
    print()
    print(ctweets_df.loc[i].text)
    print()
    
    score = input("Enter 1 for pro-Trump/anti-Biden, 0 for neutral, and -1 for pro-Biden/anti-Trump:")
    ctweets_df.loc[i, "Affiliation"] = int(score)
    tweets_df.loc[i, "Affiliation"] = int(score)
    print()
    print("--------------------------------------NEXT-TWEET:--------------------------------------")
    
```
